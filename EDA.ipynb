```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales=pd.read_csv("dirtycafe_cleaned.csv")
```

```python
sales.tail(50)
```

```python
#  In which month the cafe earn the highest earning?
# In the month of June it earn the highest revenue
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
monthly_rev = sales.groupby('Month')['Total Spent'].sum().reindex(month_order).reset_index()


sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

# Create the Line Plot (The Trend)
sns.lineplot(data=monthly_rev, x='Month', y='Total Spent', marker='o', color='royalblue', linewidth=2.5)

# Create the Bar Plot (The Volume)
sns.barplot(data=monthly_rev, x='Month', y='Total Spent', alpha=0.3, color='skyblue')

# Customizing the chart
plt.title('Monthly Revenue Trend - Cafe Sales', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.xticks(rotation=45)
```

```python
sales['Transaction Date'] = pd.to_datetime(sales['Transaction Date'])
sales['Month'] = sales['Transaction Date'].dt.month_name()

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']


monthly_trend = sales.groupby('Month')['Total Spent'].sum().reindex(month_order)


print(monthly_trend)
```

```python
# which item is sold the most
# cake is sold the most
```

```python

best_selling_items = sales.groupby('Item')['Quantity'].sum().sort_values(ascending=False)


plt.figure(figsize=(10, 6))
sns.barplot(x=best_selling_items.values, y=best_selling_items.index, palette='magma')
plt.title('Total Quantity Sold per Item')
plt.xlabel('Quantity Sold')
plt.ylabel('Item Name')
plt.show()
```

```python
# which item makes the most revenue
# salad 
```

```python
# 1. Filter out 'Unknown' and 'nan'
filtered_sales = sales[~sales['Item'].isin(['Unknown', 'nan'])]

# 2. Calculate percentages
item_revenue = filtered_sales.groupby('Item')['Total Spent'].sum().sort_values(ascending=False)
item_pct = (item_revenue / item_revenue.sum()) * 100

# 3. Create Pie Chart
plt.figure(figsize=(8,6))
colors = sns.color_palette('pastel')
explode = [0.1] + [0] * (len(item_pct) - 1) # Highlight the top item

plt.pie(item_pct, 
        labels=item_pct.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=colors,
        explode=explode, 
        shadow=True)

plt.title('Revenue Contribution by Identified Items', fontsize=15)
plt.axis('equal') 
plt.tight_layout()
plt.show()
```

```python


month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
busy_data = sales.groupby('Month').size().reindex(month_order)


plt.figure(figsize=(12, 6))
sns.barplot(x=busy_data.index, y=busy_data.values, palette='coolwarm')


plt.title('Monthly Transaction Volume (Busyness)', fontsize=16)
plt.ylabel('Total Transactions')
plt.xlabel('Month')
plt.xticks(rotation=45)

plt.show()
```
