```python
price_map = {
    'Coffee': 2,
    'Tea': 1.5,
    'Sandwich': 4,
    'Salad': 5,
    'Cake': 3,
    'Cookie': 1,
    'Smoothie': 4,
    'Juice': 3
}
```

```python
sales['Price Per Unit'] = sales['Item'].map(price_map)
```

```python
sales['Transaction ID']=sales['Transaction ID'].astype('string')
```

```python
sales['Item']=sales['Item'].astype('category')
sales['Payment Method'] = sales['Payment Method'].astype('category')
sales['Location'] = sales['Location'].astype('category')
```

```python
sales['Quantity'] = pd.to_numeric(sales['Quantity'], errors='coerce').astype('Int64')
```

```python
sales['Transaction Date'] = pd.to_datetime(sales['Transaction Date'], errors='coerce')
```

```python
sales['Total Spent'] = pd.to_numeric(sales['Total Spent'], errors='coerce').astype('Float64')
```

```python
sales['Price Per Unit'] = sales['Price Per Unit'].fillna(sales['Total Spent'] /sales['Quantity'])
```

```python
price_to_item = {
    2.0: 'Coffee',
    1.5: 'Tea',
    4.0: 'Sandwich', 
    5.0: 'Salad',
    3.0: 'Cake',     
    1.0: 'Cookie'
}
```

```python
sales['Item'] = sales['Item'].fillna(sales['Price Per Unit'].map(price_to_item))
```

```python
# List of items you want to keep
allowed_items = ['Coffee', 'Tea', 'Sandwich', 'Salad', 'Cake', 'Cookie', 'Smoothie', 'Juice']
```

```python
# This says: Keep the item if it's in my list, otherwise make it NaN
sales['Item'] = sales['Item'].where(sales['Item'].isin(allowed_items), np.nan)
```

```python
sales[sales['Quantity'].isna()]
```

```python
sales = sales.dropna(subset=['Item'])
```

```python
# correct price for the item
# convet the data type
# put vlaue in empty place at price with totak spending/quantity
# correct the item for the price
# convert the all item other than allowed item to nan
# delete all item which is nan
# there i no nan in item and price
```

```python


sales['Quantity'] = np.where(
    sales['Total Spent'].notna() & sales['Price Per Unit'].notna(),
    sales['Total Spent'] / sales['Price Per Unit'],
    np.nan
)
```
